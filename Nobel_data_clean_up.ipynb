```python
#pulling out STEM categories from the original data set to graph STEM speciic data
Stem_nobel=df[(df['Category']=='Physics') | (df['Category']=='Chemistry') | (df['Category']=='Medicine')]
Stem_nobel.head(10)
```

```python
#grouping data by category, year and gender
#creating a column called Winners to count and aggregate the data points
Stem_nobel = Stem_nobel.groupby(['Category', 'Gender','Year']).size().reset_index(name='Winners')
Stem_nobel.head(10)
```

```python
# Group by 'Year' and 'Gender', then count the number of winners
subdf = df.groupby(['Year', 'Gender',]).size().reset_index(name='Winners')
subdf.head(10)
```

```python
# To add rows of data to showcase 'Gender' datapoints even when they don't win
#Create a list of all years in the dataset
years = sorted(subdf['Year'].unique())

# Create an empty list to store the rows for the new DataFrame
new_rows = []

# Apply for all the years
for year in years:
  # fnding all the rows for that year
  rows_for_year = subdf[subdf['Year'] == year]

  # checking if there's a row for female winners
  female_row = rows_for_year[rows_for_year['Gender'] == 'female']

  # If there's no row for female winners, add a new row with null values
  if female_row.empty:
    new_rows.append({'Year': year, 'Gender': 'female', 'Winners':0 })

#checking if there's a row for male winners
  male_row = rows_for_year[rows_for_year['Gender'] == 'male']

  #If there's no row for male winners, add a new row with null values
  if male_row.empty:
    new_rows.append({'Year': year, 'Gender': 'male', 'Winners':0 })

# Adding the new rows to new_subdf
new_subdf = pd.concat([subdf, pd.DataFrame(new_rows)])

# Sort the new DataFrame by 'Year' and 'Gender' for better visualization
new_subdf = new_subdf.sort_values(['Year', 'Gender'])

# Now 'new_subdf' contains rows for all years and both genders,
# with null values where there were no female winners in a specific year.
new_subdf.head(10)
```

```python
#download newdf
new_subdf.to_csv('cleaned_count_data.csv', index=False)
```
